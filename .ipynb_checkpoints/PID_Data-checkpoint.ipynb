```python
import numpy as np
import csv as csv

%matplotlib widget
import matplotlib.pyplot as plt

path = "f:\CVSSandbox\PEP_Projects\PID\PID.csv"
f = open(path)
reader = csv.reader(f)

header = next(reader)
print(header)


time_from_file = []
first_order_open_loop = []
first_order_closed_loop = []
second_order_open_loop = []
second_order_closed_loop = []
integrator_open_loop = []
integrator_closed_loop = []
my_sin = []
noisy_first_order = []
derivative = []
prev_error = []
cur_error = []
delta_error = []
LP_filter_raw_sig = []
LP_filter_out_sig = []
LP_BW_filter_out_sig = []

for row in reader:
    time_from_file.append(float(row[0]))
    first_order_open_loop.append(float(row[1]))
    first_order_closed_loop.append(float(row[2]))
    second_order_open_loop.append(float(row[3]))
    second_order_closed_loop.append(float(row[4]))
    integrator_open_loop.append(float(row[5]))
    integrator_closed_loop.append(float(row[6]))
    my_sin.append(float(row[7]))
    noisy_first_order.append(float(row[8]))
    derivative.append(float(row[9]))
    prev_error.append(float(row[10]))
    cur_error.append(float(row[11]))
    LP_filter_raw_sig.append(float(row[12]))
    LP_filter_out_sig.append(float(row[13]))
    LP_BW_filter_out_sig.append(float(row[14]))
    # print(system_output)
delta_error = np.subtract(cur_error, prev_error)
ref = np.linspace(0.6321, 0.6321, 14998)
# plt.figure(figsize=(18,9))
plt.figure(layout="constrained")
plt.title("System Output")
# plt.subplot(2, 1, 1)
plt.grid()
ax = plt.gca()
ax.set_ylim([-2.0, 2])
plt.plot(time_from_file,ref, label='Time Constant Ref')
# plt.plot(time_from_file,first_order_open_loop, label='1st Order Open Loop')
# plt.plot(time_from_file,first_order_closed_loop,label='1st Order Closed Loop')
# plt.plot(time_from_file,second_order_open_loop,label='2nd Order Open Loop')
plt.plot(time_from_file,second_order_closed_loop,label='2nd Order Closed Loop')
plt.plot(time_from_file,derivative,label='Derivative')
plt.plot(time_from_file,prev_error,linestyle='dotted',label='Prev Error')
plt.plot(time_from_file,cur_error,linestyle='dotted',label='Curr Error')
# plt.plot(time_from_file,delta_error,label='Delta Error')
plt.legend()
# plt.plot(time_from_file,integrator_open_loop,label='Integrator Open Loop')
# plt.plot(time_from_file,integrator_closed_loop,label='Integrator Closed Loop')
# plt.subplot(2, 1, 2)
plt.figure(layout="constrained")
plt.grid()
ax = plt.gca()
ax.set_xlim([0, 0.1])
plt.plot(time_from_file,my_sin,label='Sine')
plt.figure(layout="constrained")
plt.grid()
plt.plot(time_from_file,noisy_first_order,label='Noisy')
plt.legend()
```

```python
import numpy as np
import csv as csv
import control as ct

%matplotlib widget
import matplotlib.pyplot as plt

t = np.linspace(1, 20, 1000) # time vector
G = ct.tf( [1], [0.5, 1]) # transfer function, Laplace re-presentation
G2 = ct.tf( [1, 2], [1, 1, 2]) # transfer function, Laplace re-presentation
# G2 = ct.tf( [4, 2], [8, 2.25, 2]) # transfer function, Laplace re-presentation
t, y = ct.step_response(G, t)
t, y2 = ct.step_response(G2, t)

ref = np.linspace(0.6321, 0.6321, 1000)

plt.figure(dpi=120)
# plt.title("1-st Order Transfer Function")
plt.grid()
ax = plt.gca()
ax.set_ylim([0, 2])
plt.plot(time_from_file,first_order_open_loop, label='Diff Equation 1st Order')
plt.plot(time_from_file,second_order_open_loop,label='Diff Equation 2st Order')
plt.plot(t,y,  linestyle='dotted', color='red', label='Laplace 1st Order')
plt.plot(t,y2,  linestyle='dotted', label='Laplace 2nd Order')
plt.plot(time_from_file,integrator_open_loop,label='Diff Equation Integrator')
plt.plot(t, ref, label='Time Constant Ref')
plt.legend()
```

```python
import numpy as np
import csv as csv
import control as ct

%matplotlib widget
import matplotlib.pyplot as plt

t = np.linspace(1, 20, 1000) # time vector
G = ct.tf([1], [0.5,0]) # transfer function, Laplace re-presentation
t, y = ct.step_response(G, t)

ref = np.linspace(0.6321, 0.6321, 1000)

plt.figure(dpi=120)
# plt.title("1-st Order Transfer Function")
plt.grid()
# plt.plot(time_from_file,first_order_open_loop, label='Diff Equation 1st Order')
plt.plot(time_from_file,integrator_open_loop,label='Diff Equation Integrator')
plt.plot(t,y,  linestyle='dotted', color='red', label='Laplace Integrator')
plt.plot(t, ref, label='Time Constant Ref')
plt.legend()
```
